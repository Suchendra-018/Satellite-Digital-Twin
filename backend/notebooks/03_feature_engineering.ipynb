```python
# ==========================================================
# Satellite Digital Twin
# Notebook 03 - Feature Engineering
# ==========================================================

from pathlib import Path
import pandas as pd

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

PROJECT_ROOT = Path("..").resolve()
DATASET_DIR = PROJECT_ROOT / "datasets"

df = pd.read_csv(DATASET_DIR / "train_scaled.csv")

print("=" * 70)
print("Satellite Digital Twin - Feature Engineering")
print("=" * 70)

# ----------------------------------------------------------
# Feature Columns
# ----------------------------------------------------------

FEATURE_COLUMNS = [
    "OrbitPhase (%)",
    "Sunlight (0 or 1)",
    "BusVoltage (V)",
    "BusCurrent (A)",
    "BatteryVoltage (V)",
    "BatteryTemperature (°C)",
    "BatterySOC (%)",
    "SolarVoltage (V)",
    "SolarCurrent (A)",
    "WheelRPM (RPM)",
    "WheelTemperature (°C)",
    "CPUUsage (%)",
    "CPUTemperature (°C)",
    "SignalStrength (dBm)",
    "GyroMagnitude (deg/s)",
    "Altitude (km)",
]

TARGET_COLUMN = "FaultLabel"

# ----------------------------------------------------------
# Split Features & Target
# ----------------------------------------------------------

X = df[FEATURE_COLUMNS]
y = df[TARGET_COLUMN]

print("\nFeature Matrix Shape")
print(X.shape)

print("\nTarget Shape")
print(y.shape)

print("\nSelected Features")
for feature in FEATURE_COLUMNS:
    print("-", feature)

print("\nTarget Classes")
print(sorted(y.unique()))

# ----------------------------------------------------------
# Feature Statistics
# ----------------------------------------------------------

print("\nFeature Summary")
display(X.describe())

# ----------------------------------------------------------
# Feature Validation
# ----------------------------------------------------------

print("\nChecking Missing Values")

missing = X.isnull().sum()

display(missing)

print("\nChecking Duplicate Rows")

print(X.duplicated().sum())

print("\nChecking Feature Data Types")

display(X.dtypes)

# ----------------------------------------------------------
# Target Distribution
# ----------------------------------------------------------

print("\nTarget Distribution")

display(y.value_counts())

# ----------------------------------------------------------
# Preview Engineered Dataset
# ----------------------------------------------------------

engineered_df = X.copy()

engineered_df[TARGET_COLUMN] = y

print("\nEngineered Dataset Preview")

display(engineered_df.head())

print("\nFeature Engineering Completed Successfully.")
```
