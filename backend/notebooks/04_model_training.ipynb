```python
# ==========================================================
# Satellite Digital Twin
# Notebook 04 - XGBoost Model Evaluation
# ==========================================================

from pathlib import Path
import pandas as pd
import joblib

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)

# ----------------------------------------------------------
# Project Paths
# ----------------------------------------------------------

PROJECT_ROOT = Path("..").resolve()

DATASET_DIR = PROJECT_ROOT / "datasets"
MODEL_DIR = PROJECT_ROOT / "saved_models"

print("=" * 70)
print("Satellite Digital Twin")
print("XGBoost Model Evaluation")
print("=" * 70)

print("\nProject Directory:")
print(PROJECT_ROOT)

print("\nDataset Directory:")
print(DATASET_DIR)

print("\nModel Directory:")
print(MODEL_DIR)

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

train_df = pd.read_csv(DATASET_DIR / "train_scaled.csv")
test_df = pd.read_csv(DATASET_DIR / "test_scaled.csv")

print("\nTrain Shape :", train_df.shape)
print("Test Shape  :", test_df.shape)

# ----------------------------------------------------------
# Feature Columns
# ----------------------------------------------------------

FEATURE_COLUMNS = [
    "OrbitPhase (%)",
    "Sunlight (0 or 1)",
    "BusVoltage (V)",
    "BusCurrent (A)",
    "BatteryVoltage (V)",
    "BatteryTemperature (°C)",
    "BatterySOC (%)",
    "SolarVoltage (V)",
    "SolarCurrent (A)",
    "WheelRPM (RPM)",
    "WheelTemperature (°C)",
    "CPUUsage (%)",
    "CPUTemperature (°C)",
    "SignalStrength (dBm)",
    "GyroMagnitude (deg/s)",
    "Altitude (km)",
]

TARGET_COLUMN = "FaultLabel"

# ----------------------------------------------------------
# Split Features & Labels
# ----------------------------------------------------------

X_train = train_df[FEATURE_COLUMNS]
y_train = train_df[TARGET_COLUMN]

X_test = test_df[FEATURE_COLUMNS]
y_test = test_df[TARGET_COLUMN]

print("\nTraining Samples :", len(X_train))
print("Testing Samples  :", len(X_test))

# ----------------------------------------------------------
# Load Saved Model
# ----------------------------------------------------------

model_path = MODEL_DIR / "xgboost.pkl"

print("\nLoading Model...")
print(model_path)

model = joblib.load(model_path)

print("Model Loaded Successfully!")

# ----------------------------------------------------------
# Make Predictions
# ----------------------------------------------------------

predictions = model.predict(X_test)

# ----------------------------------------------------------
# Accuracy
# ----------------------------------------------------------

accuracy = accuracy_score(y_test, predictions)

print("\nAccuracy")
print("-" * 30)
print(f"{accuracy*100:.2f}%")

# ----------------------------------------------------------
# Classification Report
# ----------------------------------------------------------

print("\nClassification Report")
print("-" * 30)

print(classification_report(
    y_test,
    predictions
))

# ----------------------------------------------------------
# Confusion Matrix
# ----------------------------------------------------------

cm = confusion_matrix(
    y_test,
    predictions
)

print("\nConfusion Matrix")
print("-" * 30)

print(cm)

# ----------------------------------------------------------
# Sample Predictions
# ----------------------------------------------------------

results = pd.DataFrame({

    "Actual": y_test.values,

    "Predicted": predictions

})

print("\nFirst 20 Predictions")

display(results.head(20))

# ----------------------------------------------------------
# Prediction Distribution
# ----------------------------------------------------------

print("\nPrediction Distribution")

display(results["Predicted"].value_counts())

# ----------------------------------------------------------
# Example Prediction
# ----------------------------------------------------------

print("\nExample Prediction")

sample = X_test.iloc[[0]]

prediction = model.predict(sample)[0]

print("Predicted Fault :", prediction)

print("\nInput Features")

display(sample)

print("\nNotebook Completed Successfully.")
```
