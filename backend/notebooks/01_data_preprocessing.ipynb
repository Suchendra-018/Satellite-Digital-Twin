```python
# ==========================================================
# Satellite Digital Twin
# Notebook 01 - Data Preprocessing
# ==========================================================

from pathlib import Path
import pandas as pd

# ----------------------------------------------------------
# Project Paths
# ----------------------------------------------------------

PROJECT_ROOT = Path("..").resolve()

DATASET_DIR = PROJECT_ROOT / "datasets"

# ----------------------------------------------------------
# Load Datasets
# ----------------------------------------------------------

train = pd.read_csv(DATASET_DIR / "train_scaled.csv")
test = pd.read_csv(DATASET_DIR / "test_scaled.csv")
demo_scaled = pd.read_csv(DATASET_DIR / "demo_scaled.csv")
demo_original = pd.read_csv(DATASET_DIR / "demo.csv")

# ----------------------------------------------------------
# Feature Columns
# ----------------------------------------------------------

FEATURE_COLUMNS = [
    "OrbitPhase (%)",
    "Sunlight (0 or 1)",
    "BusVoltage (V)",
    "BusCurrent (A)",
    "BatteryVoltage (V)",
    "BatteryTemperature (°C)",
    "BatterySOC (%)",
    "SolarVoltage (V)",
    "SolarCurrent (A)",
    "WheelRPM (RPM)",
    "WheelTemperature (°C)",
    "CPUUsage (%)",
    "CPUTemperature (°C)",
    "SignalStrength (dBm)",
    "GyroMagnitude (deg/s)",
    "Altitude (km)",
]

TARGET_COLUMN = "FaultLabel"

# ----------------------------------------------------------
# Split Features & Target
# ----------------------------------------------------------

X_train = train[FEATURE_COLUMNS]
y_train = train[TARGET_COLUMN]

X_test = test[FEATURE_COLUMNS]
y_test = test[TARGET_COLUMN]

# ----------------------------------------------------------
# Dataset Information
# ----------------------------------------------------------

print("=" * 60)
print("Satellite Digital Twin - Data Preprocessing")
print("=" * 60)

print("\nTrain Dataset :", train.shape)
print("Test Dataset  :", test.shape)
print("Demo Dataset  :", demo_original.shape)

print("\nFeatures :", len(FEATURE_COLUMNS))

print("\nTarget Column :", TARGET_COLUMN)

print("\nFault Classes")

print(sorted(train[TARGET_COLUMN].unique()))

print("\nMissing Values")

print(train.isnull().sum())

print("\nFeature Matrix")

print(X_train.shape)

print("\nTarget Vector")

print(y_train.shape)

print("\nFirst Five Samples")

display(train.head())

print("\nFeature Statistics")

display(X_train.describe())

print("\nTarget Distribution")

display(y_train.value_counts())

print("\nData Preprocessing Completed Successfully.")
```
