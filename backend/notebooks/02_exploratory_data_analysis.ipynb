```python
# ==========================================================
# Satellite Digital Twin
# Notebook 02 - Exploratory Data Analysis (EDA)
# ==========================================================

from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# Load Dataset
# ----------------------------------------------------------

PROJECT_ROOT = Path("..").resolve()
DATASET_DIR = PROJECT_ROOT / "datasets"

df = pd.read_csv(DATASET_DIR / "train_scaled.csv")

print("=" * 70)
print("Satellite Digital Twin - Exploratory Data Analysis")
print("=" * 70)

# ----------------------------------------------------------
# Basic Information
# ----------------------------------------------------------

print("\nDataset Shape")
print(df.shape)

print("\nColumn Names")
print(df.columns.tolist())

print("\nFirst Five Rows")
display(df.head())

print("\nDataset Information")
df.info()

print("\nMissing Values")
display(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())

print("\nStatistical Summary")
display(df.describe(include="all"))

# ----------------------------------------------------------
# Fault Distribution
# ----------------------------------------------------------

print("\nFault Class Distribution")

fault_counts = df["FaultLabel"].value_counts()

display(fault_counts)

plt.figure(figsize=(8,5))

fault_counts.plot(
    kind="bar",
    color="steelblue",
    edgecolor="black"
)

plt.title("Fault Class Distribution")
plt.xlabel("Fault Class")
plt.ylabel("Samples")
plt.grid(axis="y", alpha=0.3)

plt.show()

# ----------------------------------------------------------
# Numeric Features
# ----------------------------------------------------------

numeric_df = df.select_dtypes(include=["number"])

print("\nNumeric Features")

print(numeric_df.columns.tolist())

# ----------------------------------------------------------
# Correlation Matrix
# ----------------------------------------------------------

corr = numeric_df.corr()

plt.figure(figsize=(12,10))

plt.imshow(corr, cmap="viridis")

plt.colorbar()

plt.xticks(
    range(len(corr.columns)),
    corr.columns,
    rotation=90
)

plt.yticks(
    range(len(corr.columns)),
    corr.columns
)

plt.title("Correlation Matrix")

plt.tight_layout()

plt.show()

# ----------------------------------------------------------
# Feature Distribution
# ----------------------------------------------------------

features = [
    "BusVoltage (V)",
    "BatteryVoltage (V)",
    "BatteryTemperature (°C)",
    "BatterySOC (%)",
    "CPUUsage (%)",
    "CPUTemperature (°C)",
]

for feature in features:

    if feature not in df.columns:
        continue

    plt.figure(figsize=(6,4))

    plt.hist(
        df[feature],
        bins=30,
        edgecolor="black"
    )

    plt.title(feature)

    plt.xlabel(feature)

    plt.ylabel("Frequency")

    plt.grid(alpha=0.3)

    plt.show()

# ----------------------------------------------------------
# Boxplots
# ----------------------------------------------------------

for feature in features:

    if feature not in df.columns:
        continue

    plt.figure(figsize=(6,3))

    plt.boxplot(df[feature])

    plt.title(feature)

    plt.grid(alpha=0.3)

    plt.show()

print("\nEDA Completed Successfully.")
```
